```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = gk_df.select_dtypes(include=[np.number]).columns.tolist()
feature_cols = [col for col in numeric_cols if col.lower() != 'age']

print(f"Total numeric columns: {len(numeric_cols)}")
print(f"Features for PCA (excluding Age): {len(feature_cols)}")
print(f"Feature columns: {feature_cols}")

gk_df[feature_cols] = gk_df[feature_cols].fillna(gk_df[feature_cols].median())
```

```python
player_names = gk_df['Player'].values

scaler = StandardScaler()
scaled_features = scaler.fit_transform(gk_df[feature_cols])

pca_temp = PCA()
pca_temp.fit(scaled_features)
cumsum = np.cumsum(pca_temp.explained_variance_ratio_)
n_components = np.argmax(cumsum >= 0.95) + 1
print(f"Components needed for 95% variance: {n_components}")

pca = PCA(n_components=n_components)
pca_features = pca.fit_transform(scaled_features)

print(f"PCA fitted with {n_components} components")
print(f"Total variance explained: {pca.explained_variance_ratio_.sum():.4f}")
```

```python
from sklearn.neighbors import NearestNeighbors


knn = NearestNeighbors(n_neighbors=4, metric='cosine', algorithm='brute')
knn.fit(pca_features)

n_comp_viz = min(3, pca.n_components_)
fig, axes = plt.subplots(1, n_comp_viz, figsize=(15, 5))

for i in range(n_comp_viz):
    axes[i].hist(pca_features[:, i], bins=50, edgecolor='black', alpha=0.7, color='steelblue')
    axes[i].set_xlabel(f'PC{i+1} (Var: {pca.explained_variance_ratio_[i]:.3f})', fontsize=12)
    axes[i].set_ylabel('Frequency', fontsize=12)
    axes[i].set_title(f'Distribution of PC{i+1}', fontsize=14, fontweight='bold')
    axes[i].grid(alpha=0.3)

plt.tight_layout()
plt.show()

fig = plt.figure(figsize=(12, 10))

ax1 = plt.subplot(2, 2, 1)
scatter = ax1.scatter(pca_features[:, 0], pca_features[:, 1], 
                    alpha=0.6, c=range(len(pca_features)), cmap='viridis', s=30)
ax1.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.3f})', fontsize=11)
ax1.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.3f})', fontsize=11)
ax1.set_title('PCA: PC1 vs PC2', fontsize=13, fontweight='bold')
ax1.grid(alpha=0.3)

ax2 = plt.subplot(2, 2, 2)
cumsum_plot = np.cumsum(pca.explained_variance_ratio_)
ax2.plot(range(1, len(cumsum_plot)+1), cumsum_plot, marker='o', linewidth=2, markersize=6, color='red')
ax2.axhline(y=0.95, color='green', linestyle='--', label='95% Variance', linewidth=2)
ax2.set_xlabel('Number of Components', fontsize=11)
ax2.set_ylabel('Cumulative Variance Explained', fontsize=11)
ax2.set_title('Cumulative Variance Explained', fontsize=13, fontweight='bold')
ax2.legend()
ax2.grid(alpha=0.3)


ax3 = plt.subplot(2, 2, 3)
ax3.scatter(pca_features[:, 0], pca_features[:, 2], 
           alpha=0.6, c=range(len(pca_features)), cmap='plasma', s=30)
ax3.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.3f})', fontsize=11)
ax3.set_ylabel(f'PC3 ({pca.explained_variance_ratio_[2]:.3f})', fontsize=11)
ax3.set_title('PCA: PC1 vs PC3', fontsize=13, fontweight='bold')
ax3.grid(alpha=0.3)

ax4 = plt.subplot(2, 2, 4)
loadings = pca.components_[:2, :].T
top_features = 10
feature_importance = np.abs(loadings).sum(axis=1)
top_indices = np.argsort(feature_importance)[-top_features:]

y_pos = np.arange(top_features)
ax4.barh(y_pos, feature_importance[top_indices], color='teal', alpha=0.7)
ax4.set_yticks(y_pos)
ax4.set_yticklabels([feature_cols[i][:20] for i in top_indices], fontsize=9)
ax4.set_xlabel('Total Absolute Loading', fontsize=11)
ax4.set_title(f'Top {top_features} Feature Contributions', fontsize=13, fontweight='bold')
ax4.grid(alpha=0.3, axis='x')

plt.tight_layout()
plt.show()
```

```python
player_to_search = "Thibaut Courtois"

player_mask = player_names == player_to_search
if not player_mask.any():
    player_mask = np.char.lower(player_names.astype(str)) == player_to_search.lower()
    if not player_mask.any():
        print(f"Player '{player_to_search}' not found in dataset.")
        print(f"Sample player names: {player_names[:5]}")
    else:
        player_idx = np.where(player_mask)[0][0]
else:
    player_idx = np.where(player_mask)[0][0]

if player_mask.any():
    player_pca = pca_features[player_idx].reshape(1, -1)
    
    distances, indices = knn.kneighbors(player_pca, n_neighbors=4)
    
    distances = distances[0][1:4]
    indices = indices[0][1:4]
    
    results = []
    for idx, dist in zip(indices, distances):
        player_info = {
            'Name': player_names[idx],
            'Distance': dist,
        }
        
        for col in ['Player', 'Nation', 'Pos', 'Squad', 'Comp', 'Age']:
            if col in gk_df.columns:
                player_info[col] = gk_df.iloc[idx][col]
                
        results.append(player_info)
    
    results_gk_df = pd.DataFrame(results)
    
    # Professional formatted output
    print("\n" + "â" * 90)
    print(f"\t\t\t\tPLAYER SIMILARITY ANALYSIS")
    print("â" * 90 + "\n")
    
    print(f"  Target Player: {player_to_search}")
    
    player_info_items = []
    for col in ['Nation', 'Pos', 'Squad', 'Comp', 'Age']:
        if col in gk_df.columns:
            value = gk_df.iloc[player_idx][col]
            player_info_items.append(f"{col}: {value}")
    
    if player_info_items:
        print(f"  {' â '.join(player_info_items)}")
    
    print("\n" + ("â" * 90 + "\n"))
    print("\t\t\t  TOP 3 MOST SIMILAR PLAYERS\n")
    
    from tabulate import tabulate
    
    table_data = []
    for i, (idx, dist) in enumerate(zip(indices, distances), 1):
        row = [f"#{i}", player_names[idx]]
        
        if 'Nation' in gk_df.columns:
            row.append(gk_df.iloc[idx]['Nation'])
        if 'Age' in gk_df.columns:
            age = gk_df.iloc[idx]['Age']
            row.append(f"{int(age)}" if pd.notna(age) else "N/A")
        if 'Squad' in gk_df.columns:
            row.append(gk_df.iloc[idx]['Squad'])
        
        row.append(f"{dist:.3f}")
        table_data.append(row)
    
    headers = ["Rank", "Player", "Nation", "Age", "Club", "Similarity"]
    
    print(tabulate(table_data, headers=headers, tablefmt="rounded_grid", 
                   stralign="left", numalign="right"))


    print("\n\n\n\n")
    similar_indices = indices
    
    fig, ax = plt.subplots(figsize=(16, 10))
    fig.patch.set_facecolor('#0d1117')
    ax.set_facecolor('#0d1117')
    
    ax.scatter(pca_features[:, 0], pca_features[:, 1], 
              alpha=0.15, c='#6e7681', s=80, label='All Players', edgecolors='none')
    
    for idx in similar_indices:
        ax.plot([pca_features[player_idx, 0], pca_features[idx, 0]], 
               [pca_features[player_idx, 1], pca_features[idx, 1]], 
               color='#58a6ff', alpha=0.4, linewidth=3, zorder=1)
    
    ax.scatter(pca_features[similar_indices, 0], 
              pca_features[similar_indices, 1],
              alpha=0.9, c='#58a6ff', s=600, marker='o', 
              label='Similar Players', edgecolors='white', linewidth=2.5, zorder=3)
    
    ax.scatter(pca_features[player_idx, 0], 
              pca_features[player_idx, 1],
              alpha=1.0, c='#f85149', s=900, marker='*', 
              label=f'{player_to_search}', edgecolors='white', linewidth=3, zorder=4)
    
    ax.annotate(player_to_search, 
               (pca_features[player_idx, 0], pca_features[player_idx, 1]),
               xytext=(15, 15), textcoords='offset points', fontsize=11, fontweight='bold',
               color='#f85149',
               bbox=dict(boxstyle='round,pad=0.5', facecolor='#0d1117', alpha=0.8, 
                        edgecolor='#f85149', linewidth=1.5),
               arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.3', 
                              color='#f85149', lw=2))
    
    for i, idx in enumerate(similar_indices):
        # Calculate offset direction
        offset_x = (pca_features[idx, 0] - pca_features[player_idx, 0]) * 0.08
        offset_y = (pca_features[idx, 1] - pca_features[player_idx, 1]) * 0.08
        
        stagger_x = (i % 3 - 1) * 0.05
        stagger_y = (i % 2) * 0.05
        
        name = player_names[idx][:15] + '...' if len(player_names[idx]) > 15 else player_names[idx]
        
        ax.annotate(f"{name}\nDist: {distances[i]:.2f}", 
                   (pca_features[idx, 0], pca_features[idx, 1]),
                   xytext=(10 + offset_x + stagger_x, 10 + offset_y + stagger_y), 
                   textcoords='offset points', fontsize=9, fontweight='bold',
                   color='#58a6ff',
                   bbox=dict(boxstyle='round,pad=0.4', facecolor='#0d1117', alpha=0.7, 
                            edgecolor='#58a6ff', linewidth=1),
                   arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.2', 
                                  color='#58a6ff', lw=1.5))
    
    ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%})', 
                 fontsize=12, fontweight='bold', color='#c9d1d9', labelpad=10)
    ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%})', 
                 fontsize=12, fontweight='bold', color='#c9d1d9', labelpad=10)
    ax.set_title(f'Player Similarity Map: {player_to_search}', 
                fontsize=16, fontweight='bold', color='#c9d1d9', pad=20)
    
    legend = ax.legend(loc='upper right', fontsize=10, framealpha=0.8, 
                      facecolor='#0d1117', edgecolor='#30363d', labelcolor='#c9d1d9', 
                      frameon=True)
    legend.get_frame().set_linewidth(1.5)
    
    ax.grid(True, alpha=0.7, color='#30363d', linestyle='--', linewidth=0.5)
    
    ax.tick_params(colors='#8b949e', labelsize=10)
    
    for spine in ax.spines.values():
        spine.set_edgecolor('#30363d')
        spine.set_linewidth(1.5)
    
    x_range = pca_features[:, 0].max() - pca_features[:, 0].min()
    y_range = pca_features[:, 1].max() - pca_features[:, 1].min()
    
    ax.set_xlim([pca_features[:, 0].min() - x_range*0.1, 
                 pca_features[:, 0].max() + x_range*0.1])
    ax.set_ylim([pca_features[:, 1].min() - y_range*0.1, 
                 pca_features[:, 1].max() + y_range*0.1])
    
    plt.tight_layout()
    plt.savefig(f'similarity_space_{player_to_search.replace(" ", "_")}2D.png', 
                dpi=300, facecolor='#0d1117', bbox_inches='tight')
    plt.show()
  
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = of_df.select_dtypes(include=[np.number]).columns.tolist()
feature_cols = [col for col in numeric_cols if col.lower() != 'age']

print(f"Total numeric columns: {len(numeric_cols)}")
print(f"Features for PCA (excluding Age): {len(feature_cols)}")
print(f"Feature columns: {feature_cols}")

of_df[feature_cols] = of_df[feature_cols].fillna(of_df[feature_cols].median())
```

```python
player_names = of_df['Player'].values

scaler = StandardScaler()
scaled_features = scaler.fit_transform(of_df[feature_cols])

pca_temp = PCA()
pca_temp.fit(scaled_features)
cumsum = np.cumsum(pca_temp.explained_variance_ratio_)
n_components = np.argmax(cumsum >= 0.95) + 1
print(f"Components needed for 95% variance: {n_components}")

pca = PCA(n_components=n_components)
pca_features = pca.fit_transform(scaled_features)

print(f"PCA fitted with {n_components} components")
print(f"Total variance explained: {pca.explained_variance_ratio_.sum():.4f}")
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(n_neighbors=4, metric='cosine', algorithm='auto')
knn.fit(pca_features)

n_comp_viz = min(3, pca.n_components_)
fig, axes = plt.subplots(1, n_comp_viz, figsize=(15, 5))

for i in range(n_comp_viz):
    axes[i].hist(pca_features[:, i], bins=50, edgecolor='black', alpha=0.7, color='steelblue')
    axes[i].set_xlabel(f'PC{i+1} (Var: {pca.explained_variance_ratio_[i]:.3f})', fontsize=12)
    axes[i].set_ylabel('Frequency', fontsize=12)
    axes[i].set_title(f'Distribution of PC{i+1}', fontsize=14, fontweight='bold')
    axes[i].grid(alpha=0.3)

plt.tight_layout()
plt.show()

fig = plt.figure(figsize=(12, 10))

ax1 = plt.subplot(2, 2, 1)
scatter = ax1.scatter(pca_features[:, 0], pca_features[:, 1], 
                    alpha=0.6, c=range(len(pca_features)), cmap='viridis', s=30)
ax1.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.3f})', fontsize=11)
ax1.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.3f})', fontsize=11)
ax1.set_title('PCA: PC1 vs PC2', fontsize=13, fontweight='bold')
ax1.grid(alpha=0.3)

ax2 = plt.subplot(2, 2, 2)
cumsum_plot = np.cumsum(pca.explained_variance_ratio_)
ax2.plot(range(1, len(cumsum_plot)+1), cumsum_plot, marker='o', linewidth=2, markersize=6, color='red')
ax2.axhline(y=0.95, color='green', linestyle='--', label='95% Variance', linewidth=2)
ax2.set_xlabel('Number of Components', fontsize=11)
ax2.set_ylabel('Cumulative Variance Explained', fontsize=11)
ax2.set_title('Cumulative Variance Explained', fontsize=13, fontweight='bold')
ax2.legend()
ax2.grid(alpha=0.3)


ax3 = plt.subplot(2, 2, 3)
ax3.scatter(pca_features[:, 0], pca_features[:, 2], 
           alpha=0.6, c=range(len(pca_features)), cmap='plasma', s=30)
ax3.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.3f})', fontsize=11)
ax3.set_ylabel(f'PC3 ({pca.explained_variance_ratio_[2]:.3f})', fontsize=11)
ax3.set_title('PCA: PC1 vs PC3', fontsize=13, fontweight='bold')
ax3.grid(alpha=0.3)

ax4 = plt.subplot(2, 2, 4)
loadings = pca.components_[:2, :].T
top_features = 10
feature_importance = np.abs(loadings).sum(axis=1)
top_indices = np.argsort(feature_importance)[-top_features:]

y_pos = np.arange(top_features)
ax4.barh(y_pos, feature_importance[top_indices], color='teal', alpha=0.7)
ax4.set_yticks(y_pos)
ax4.set_yticklabels([feature_cols[i][:20] for i in top_indices], fontsize=9)
ax4.set_xlabel('Total Absolute Loading', fontsize=11)
ax4.set_title(f'Top {top_features} Feature Contributions', fontsize=13, fontweight='bold')
ax4.grid(alpha=0.3, axis='x')

plt.tight_layout()
plt.show()
```
